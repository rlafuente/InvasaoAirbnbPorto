```python
import pandas as pd
import numpy as np
import sqlite3
import re
from tqdm import tqdm

df = pd.read_csv('joined_output.csv')
dfCodigo = pd.read_csv('codigos_postais.csv')

def codigo(row):
    if row['ext_cod_postal'] > 99:
        return str(row['num_cod_postal'])+'-'+str(row['ext_cod_postal'])
    else:
        return str(row['num_cod_postal'])+'-0'+str(row['ext_cod_postal'])
    
    
dfCodigo['codigototal'] = dfCodigo.apply(lambda row: codigo(row), axis = 1)

dfCodigo.head()

```
